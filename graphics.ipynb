```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("metricas.csv")
df
```

```python
df.groupby("procesos")["user_avg_s"].mean().plot(kind="bar")
```

```python
sns.boxplot(data=df, x='procesos', y='sys_avg_s')
plt.show()
```

```python
resumen = df.groupby('procesos')['wall_avg_s'].agg(['mean', 'std']).reset_index()

plt.errorbar(
    resumen['procesos'],
    resumen['mean'],
    yerr=resumen['std'],
    fmt='--*',      # línea con puntos
    capsize=5,      # tamaño de las "tapitas"
)

plt.xlabel('procesos')
plt.ylabel('wall_avg_s')
plt.title('Media ± Desviación estándar')
plt.grid(True)

plt.show()
```
